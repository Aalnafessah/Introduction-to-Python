# What is Coding?

## Why Code?
Writing and using custom code is an integral part of part of modern research. Simulation is a common tool use for predicting behaviour quicker and more cheaply and easily that an experiment. Even for experimental programs, custom code can be used to collect, analyse and display results. Code allows us to:
- Automate laborious calculations
- Predict behaviour
- Rapidly iterate over a number of related cases
- Can keep up with fast real-world behaviour and data gathering

## What do we Mean by "Coding"?
"Coding" is the process of writing human-readable instructions (also known as "source code") which will be translated into machine-readable instructions (also known as "machine code") which can be interpreted by the processor of the computer.

You will almost always be interacting with the source code. In this you can specify a set of instructions you want the computer to perform, using the commands available in the program language you are coding with. The things you may ask the computer to do and the syntax you used to do it will vary from programming language to programming language. 

However, source code may be written in different languages but lead to the machine-code that has the same effect. In this sense, different coding lanaguges act as different interfaces between the coder and their machine, each having their own strengths, weaknesses and quirks. Some examples of coding languages include:
- Python
- C
- C++
- Fortran
- Julia
- Javascript

The list of programming languages is very long indeed and includes many related families of languages. In many cases, a similar logic is used to construct the program and this means that, once you know your first language, learning your second will be much easier.

In order to code in a language effectively, you must:
- Know which operations are available in the language you're using
- Be able to work out which series of these operations which achieves your goal
- Know the syntax of the language and use it to create a source file which passes these instructions to the computer
- Know how to instruct the computer to execute your list of instructions

This course aims to give you basic abilities in all four of these capabilities.
